### Now deciding to work with 'PlatformsHaveWorkedWith' and 'YearsCode' to get an idea of what platforms are used 


Sure, let's visualize each step with a simplified example. Assume we start with a DataFrame like this:

ResponseId	PlatformHaveWorkedWith
1	PlatformA;PlatformB
2	PlatformB;PlatformC
Step 1 & 2: str.split(';', expand=True).stack()
We split the 'PlatformHaveWorkedWith' column and then stack it:

ResponseId	
1	PlatformA
1	PlatformB
2	PlatformB
2	PlatformC
Step 3: index.droplevel(-1)
After stacking, the DataFrame gets a multi-level index. We drop the innermost level of this multi-index:

Before droplevel:

Index structure: (ResponseId, [some auto-generated index])
Example: (1, 0), (1, 1), (2, 0), (2, 1)
After droplevel:

Index structure: (ResponseId)
Example: 1, 1, 2, 2
Step 4: df_platforms.name = 'Platform'
We name the Series created by stack() as 'Platform', which is essentially the column name for our stacked data.

ResponseId	Platform
1	PlatformA
1	PlatformB
2	PlatformB
2	PlatformC
Step 5: df.join(df_platforms)
We join this back to the original DataFrame. The join is done on the index, so the 'Platform' column aligns with the corresponding 'ResponseId':

ResponseId	PlatformHaveWorkedWith	Platform
1	PlatformA;PlatformB	PlatformA
1	PlatformA;PlatformB	PlatformB
2	PlatformB;PlatformC	PlatformB
2	PlatformB;PlatformC	PlatformC
After this process, each row represents a single response-platform pair, making it easier to analyze individual platforms per respondent.

```python
#First convert 'YearsCode' to numeric
df_top_countries['YearsCode'] = pd.to_numeric(df_top_countries['YearsCode'], errors='coerce')

#Since 'PlatformHaveWorkedWith' will have multiple platforms for each entry, we split them ";" and then call .stack() to reshape/pivot 
#The stack() method is used to "pivot" or reshape the DataFrame from a wide format (where each platform is in a separate column) to a long format (where all platforms are in a single column but multiple rows).
df_platforms = df_top_countries['PlatformHaveWorkedWith'].str.split(";", expand=True).stack()
df_platforms.index = df_platforms.index.droplevel(-1)
df_platforms.name = 'Platform'
df_with_platforms = df_top_countries.join(df_platforms)
```

#### Platform Popularity Analysis

```python
#Now that we have the df_with_platforms
#Create a countplot via seaborn
popularity_plot = sns.countplot(data=df_with_platforms, y='Platform', order=df_with_platforms['Platform'].value_counts().index)
popularity_plot.set_title('Popularity of platforms worked with')
plt.show()
```

#### Analyse the Relationship between YearsCoding and platforms used 

```python
plt.figure(figsize=(10,6))
violinplot = sns.violinplot(data=df_with_platforms, x='Platform', y='YearsCode')
violinplot.set_title('Years of coding and Platforms used')
plt.xticks(rotation=90)
plt.show()
```

The violin plot provides a visual representation of the distribution of a numerical variable (in this case, 'YearsCode') for each category (the platforms) in your dataset. Here's how to interpret the plot, focusing on AWS and Heroku as examples:

Thickness of the Violin: If Heroku's violin is thicker in the 0-10 years range compared to AWS, it suggests that a larger proportion of respondents who have worked with Heroku have between 0 and 10 years of coding experience. This implies Heroku is relatively more popular among developers with less experience.

Median Line: The white dot (median) indicates the middle value of 'YearsCode' for respondents who have worked with that platform. If AWS's median is higher than Heroku's, it suggests that the typical AWS user in your survey has more coding experience than the typical Heroku user.

Range of the Violin: The full range of the violin indicates the spread of 'YearsCode' among the respondents for that platform. A long range for AWS would mean that its users have a wide range of experience, from beginners to highly experienced developers.

Comparing AWS and Heroku:

If AWS's violin has a broader width at higher 'YearsCode' values, it might indicate that AWS is popular among more experienced developers.
If Heroku's violin is broader at the lower end of 'YearsCode', it might suggest that it's a platform that attracts or is more accessible to newer developers.
It's important to note that this plot shows the distribution of experience levels for each platform but doesn't indicate the total number of respondents who use each platform. A platform could be very popular among both new and experienced developers but still have a thinner violin plot if fewer survey respondents mentioned using it.







#### Analyse relationship between YearsCodePro and Compensation


```python
import numpy as np
df_top_countries['YearsCode'] = pd.to_numeric(df_top_countries['YearsCode'], errors='coerce')
df_top_countries['YearsCodePro'] = pd.to_numeric(df_top_countries['YearsCodePro'], errors='coerce')
df_top_countries['CompTotal'] = pd.to_numeric(df_top_countries['CompTotal'], errors='coerce')

#Create a scatterplot
#Apply a log transformation on the Compensation
df_top_countries['log_CompTotal'] = np.log1p(df_top_countries['CompTotal'])
yearsCode_compensation = sns.scatterplot(data=df_top_countries, x='YearsCodePro', y='log_CompTotal')
yearsCode_compensation.set_xlabel("Years of Pro coding")
yearsCode_compensation.set_ylabel("Total Compensation")
yearsCode_compensation.set_title("Distribution of Compensation by Coding Experience in 2023")
plt.show()
```

#### Applying the IterativeImputer on YearsCodePro and CompTotal

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.model_selection import train_test_split

impute_df = df_top_countries[['YearsCodePro', 'CompTotal']]
imp = IterativeImputer(max_iter=20, random_state=0)

#Next fit the model to the data
imp.fit(impute_df)
#Now, transform the data using the fitted imputer
imputed_df = pd.DataFrame(np.round(imp.transform(impute_df), 0), columns=['YearsCodePro', 'CompTotal'])

```

```python
#Filtering data and Further Analysis
df_imputed_plot = imputed_df.loc[imputed_df['CompTotal'] <= 200000].copy()
#Create Bins Based on Years of Professional Coding Experience
df_imputed_plot['CodeYearBins'] = pd.qcut(df_imputed_plot['YearsCodePro'], q=5)
#Now create a boxplot
sns.boxplot(data=df_imputed_plot, x='CodeYearBins', y='CompTotal')
```

X-axis (CodeYearBins): The x-axis shows bins of 'YearsCodePro', which represents ranges of years of professional coding experience. Each bin groups respondents by their years of professional experience.

Y-axis (CompTotal): The y-axis shows the total compensation for respondents within each experience bin.

Box Plot Components:

The line in the middle of each box is the median compensation for each experience bin.
The box itself represents the interquartile range (IQR), with the bottom of the box at the first quartile (25th percentile) and the top of the box at the third quartile (75th percentile). This indicates where the middle 50% of compensation data falls.
The whiskers (lines extending from the top and bottom of each box) typically extend to 1.5 * IQR above the third quartile and below the first quartile. Points beyond the whiskers can be considered outliers.
Outliers are the individual points that lie beyond the whiskers, representing compensation figures that are unusually high or low compared to the rest of the data in that bin.
Interpreting the Trends:

From the plot, it seems that there's an upward trend in median compensation as years of professional coding experience increases. This is indicated by the median lines within the boxes generally moving upward as we move from left to right across the bins of 'YearsCodePro'.
The range of compensation (indicated by the IQR and the whiskers) seems to be larger for bins with more years of experience. This could imply that there's greater variability in compensation among more experienced professionals.
It's also noticeable that there are outliers across all experience levels, but especially at higher levels of experience, which suggests that there are some individuals with very high compensation relative to their peers.
Considerations:

The bins are created based on quantiles, which means each bin should contain approximately the same number of respondents. This is good for ensuring that each bin is comparably sized but means the range of years within each bin can vary.
Given that the data was imputed, it's important to remember that the values include estimates for missing data and not just observed values. This could affect the accuracy of the trends and distributions you see.
The overall trend suggested by the plot is in line with what one might expect — as professional coding experience increases, so does compensation. However, the variability and outliers indicate that experience is not the only factor affecting compensation; other factors likely play a significant role as well.
